#Discuss Challenges with GPT-2

A big problem with text generation in general that has surfaced a lot with the release of ChatGPT is that text generation can often be confidently wrong. Meaning it will generate text that sounds and looks very well written and factual, but can be inherently wrong. Also what the language model will generate heavily depends on what data it was trained on. In the case of ChatGPT it very often generates unnecesarily long and wordy statements to simple questions, which goes back to the type of training data it has had access to where probably longer sentences mean better performance. This can be dangerous if people would use it as a unmoderated source for generating for example news articles or similar things as in the case of ChatGPT it can in many cases perform very well, but in some cases its just completely wrong but will still answer with 100% confidence which can make it hard to discern whats correct and not from its answers.

[Robert Miles makes an interesting note](https://www.youtube.com/watch?v=w65p_IIp6JY) on this regard saying that AI/language models aren't actually trying to "tell the truth" and is instead just trying to say what we want it to say based on the training data. For example, what happens if we break a mirror? the language model may respond with "7 years of bad luck" - its not really the case that the language model believes that, but that the biases and text present in the training data would point to that. And in some cases, the language model may even "lie" as to say the things that it thinks we want it to say (for example in the case if we incorporate reinforcment learning and give negative and positive rewards to the agent). 

There's also the inherent biases that can be created from the training data. All a language model does is trying to predict what the next token is based on the training it went through with the available training data. So lets say we train a language model based on a huge dataset on disasters or propaganda in the news, it may have some very different answers and views on things than a model trained on for example wikipedia. Therefore it's important to recognize what biases are inherent in the training data, because the model will reflect the biases that exists in the training data we give it.

A common theme for above is that it all concerns the data in some way, so a critical part of building a strong and representative language model is to have strong and representative data that the model can learn from. This is of course way easier said than done and is one of the hardest parts of launching a successful model.
