```python
import pandas as pd
import numpy as np
import spacy

nlp =spacy.load('en_core_web_md')

#Reading the file located in my PC local files.
df = pd.read_csv('IMDB Dataset.csv')

#Removing special characters
df['review'] = df['review'].str.replace('\W', ' ', regex=True)

#Creating two DF, one for positives and one for negative.
dfpositive = df.loc[df['sentiment'] == 'positive']
dfnegative = df.loc[df['sentiment'] == 'negative']

#only extracting the column "review" from document.
dfP =dfpositive.review
dfN =dfnegative.review

#only extracting first 500 rows
dfP =dfP.head(20)
dfN =dfN.head(20)


#Making it a array and assigning to NLP 
arr = dfP.to_numpy()
doc1 = nlp(arr[0])


#creating variables for loop
i = 1
x = len(arr)

while i < x:

    doc2 =nlp(arr[i])
    print(doc1.similarity(doc2))
    print(doc2)
    i += 1
    
print(doc1)
```
